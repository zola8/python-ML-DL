````python
import pandas as pd

df = pd.read_csv('../../data/diabetes.csv')
X = df.drop(columns=['Outcome'])
````

````python
X.head()
````

**Absolute Maximum Scaling** is a feature scaling method where each value is divided by the maximum absolute value of that feature.
This transformation rescales the data so that values fall within the range of −1 to 1.

**Min-Max Scaling** rescales features by subtracting the minimum value and dividing by the difference between the maximum and minimum values. This usually maps feature values to the range 0 to 1 while preserving the original distribution.

**Normalization** scales each data sample so that its vector length (Euclidean norm) becomes 1.
It focuses on the direction of data points rather than their magnitude, making it useful in tasks like text classification and clustering.

````python
# MinMaxScaler: maps feature values to the range 0 to 1 while preserving the original distribution.
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(X)
scaled_df = pd.DataFrame(scaled_data, columns=X.columns)

scaled_df.head()
````

````python
# Standardization: This transforms the data so that features have zero mean and unit variance

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(X)
scaled_df = pd.DataFrame(scaled_data, columns=X.columns)

scaled_df.head()
````

````python
# Normalization scales each data sample so that its vector length (Euclidean norm) becomes 1.
# It focuses on the direction of data points rather than their magnitude, making it useful in tasks like text classification and clustering.

from sklearn.preprocessing import Normalizer

scaler = Normalizer()
scaled_data = scaler.fit_transform(X)
scaled_df = pd.DataFrame(scaled_data, columns=X.columns)

scaled_df.head()
````

````python
# Robust Scaling scales features using the median and interquartile range (IQR) instead of the mean and standard deviation.
# This makes it less sensitive to outliers and skewed data, making it suitable for datasets with extreme values or noise.

from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
scaled_data = scaler.fit_transform(X)
scaled_df = pd.DataFrame(scaled_data, columns=X.columns)

scaled_df.head()
````

````python
# Scaling selected column

scaler = MinMaxScaler()
selected_cols = ['Pregnancies', 'Glucose']
scaled_data = scaler.fit_transform(X[selected_cols])
scaled_df = pd.DataFrame(scaled_data, columns=selected_cols)

X[selected_cols] = scaled_df[selected_cols]
X.head()
````
