### 5. Checking Duplicates Data

```python
print(f"Duplicates in Train Dataset is: {train_df.duplicated().sum()} ({100 * train_df.duplicated().sum() / len(train_df)}%)")
```

```python
row_to_duplicate = train_df.iloc[[-1]]
df_new = pd.concat([train_df, row_to_duplicate], ignore_index=False).reset_index(drop=True)
```

```python
print(f"Duplicates in the new Dataset is: {df_new.duplicated().sum()} ({100 * df_new.duplicated().sum() / len(df_new)}%)")
```

```python
# duplicated showing after the 2nd occurrence!

df_new[df_new.duplicated()]
```

```python
df_new[df_new.duplicated(keep=False)]
```

```python
# This will show the ORIGINAL row (the first one)

df_new[df_new.duplicated(keep='last')]
```

```python
df_new = df_new.drop_duplicates()
```

```python
print("Number of duplicates:", df_new.duplicated().sum())
```
