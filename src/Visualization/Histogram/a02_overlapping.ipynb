```python
import numpy as np
from matplotlib import pyplot as plt

data1 = np.random.normal(100, 15, 1000)  # Mean=100, Std=15, 1000 points
data2 = np.random.normal(120, 20, 1000)

plt.figure(figsize=(10, 6))
plt.hist(data1, bins=30, color='blue', alpha=0.6, label='Dataset 1')
plt.hist(data2, bins=30, color='red', alpha=0.6, label='Dataset 2')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Overlapping Histograms')
plt.grid(True, axis='y', alpha=0.3)
plt.show()
```
