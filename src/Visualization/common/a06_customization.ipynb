```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

plt.style.use('default')
np.random.seed(234)

x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.sin(x)
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(x, y, 'b-', linewidth=2, label='sin(x)')

# add text annotations
max_idx = np.argmax(y)
ax.annotate('Maximum',
            xy=(x[max_idx], y[max_idx]),
            xytext=(x[max_idx] + 0.5, y[max_idx] + 0.2),
            arrowprops=dict(color='red', lw=2, arrowstyle='->'),
            fontsize=12, color='red', fontweight='bold'
            )

# add text box
ax.text(0.5, 0.95, 'This is a sine wave',
        transform=ax.transAxes,
        fontsize=12,
        verticalalignment='top',
        bbox=dict(boxstyle='round', fc='wheat', alpha=0.5),
        )

# add vertical and horizontal line
ax.axvline(x=5, color='green', linestyle='--', alpha=0.5, label='x=5')
ax.axhline(y=0, color='black', linestyle='-', alpha=0.3)

ax.set_xlabel('x values', fontsize=12)
ax.set_ylabel('sin(x)', fontsize=12)
ax.set_title("Annotated Plot", fontsize=14, fontweight='bold')
ax.legend(loc='upper right', fontsize=12, frameon=True, fancybox=False, shadow=True, title="Functions")
ax.grid(True, alpha=0.3)
plt.show()
```
