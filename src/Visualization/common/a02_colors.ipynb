```python
import math

import matplotlib.colors as mcolors
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
```

```python
def plot_colors(colors):
    fig, ax = plt.subplots(figsize=(4, len(colors) * 0.4))

    for i, (name, code) in enumerate(colors.items()):
        y = len(colors) - i - 1
        ax.add_patch(Rectangle((0, y), 1, 0.8, facecolor=code, edgecolor="black"))
        ax.text(1.1, y + 0.4, f"{name}  {code}", va="center")

    ax.set_xlim(0, 3)
    ax.set_ylim(0, len(colors))
    ax.axis("off")

    plt.show()
```

```python
plot_colors(mcolors.CSS4_COLORS)
```
