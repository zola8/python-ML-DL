```python
import matplotlib.pyplot as plt
```

```python
# Sample data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [1500, 1800, 1700, 1900, 2100, 2300, 2500, 2400, 2200, 2000, 1900, 2100]
```

```python
# Create a line chart
plt.figure(figsize=(10, 5))
plt.plot(months, sales, marker='o', linestyle='-', color='blue')
plt.title('Monthly Sales Trend (Matplotlib)')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Seaborn visualization
import seaborn as sns
import pandas as pd

# Sample data for Seaborn
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Sales': [1500, 1800, 1700, 1900, 2100, 2300, 2500, 2400, 2200, 2000, 1900, 2100]
}
df = pd.DataFrame(data)

# Create a line chart with Seaborn
plt.figure(figsize=(10, 5))
sns.lineplot(x='Month', y='Sales', data=df, marker='o', color='b')
plt.title('Monthly Sales Trend (Seaborn)')
plt.grid(True, alpha=0.3)
plt.show()
```
