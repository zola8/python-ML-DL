```python
import numpy as np
from matplotlib import pyplot as plt

# grouped bar chart

categories = ['Q1', 'Q2', 'Q3', 'Q4']
product_a = [20, 35, 30, 35]
product_b = [25, 32, 34, 20]
product_c = [30, 25, 30, 25]

x = np.arange(len(categories))
width = 0.25  # width of bar

plt.figure(figsize=(10, 6))
plt.bar(x - width, product_a, width, color='#B27456', edgecolor='black', linewidth=1, label='Product A')
plt.bar(x, product_b, width, color='#542256', edgecolor='black', linewidth=1, label='Product B')
plt.bar(x + width, product_c, width, color='#7425A7', edgecolor='black', linewidth=1, label='Product C')

plt.xlabel("Quarter")
plt.ylabel("Sales")
plt.title("Quarterly Sales Comparison")
plt.grid(True, alpha=0.2, axis='y')
plt.xticks(x, categories)
plt.legend()
plt.show()
```
