```python
import matplotlib.pyplot as plt

# Sample data
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 24, 36, 48, 50]

# Create a bar chart
plt.figure(figsize=(10, 5))
plt.bar(categories, values, color='c')
plt.title('Category Values')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

```python
import pandas as pd

data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Sales': [1500, 1800, 1700, 1900, 2100, 2300, 2500, 2400, 2200, 2000, 1900, 2100]
}
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(10, 5))
plt.bar(df['Month'], df['Sales'], color='goldenrod')
plt.title('Category Values')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```
