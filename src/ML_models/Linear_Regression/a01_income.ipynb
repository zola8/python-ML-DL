```python
import pandas as pd

df = pd.read_csv('../../data/income.csv', index_col=0)

df[:3]
```

```python
from matplotlib import pyplot as plt

# Show the Relationship between the income and the happiness

plt.scatter(df.income, df.happiness)
plt.title("The Correlation between Income and Happiness")
plt.ylabel("Happiness")
plt.xlabel("Income")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(df[['income']], df['happiness'])
```

```python
y_pred = model.predict(df[['income']])
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df[['income']], df['happiness'], color='blue', label='Data Points')
plt.plot(df[['income']], y_pred, color='red', linewidth=2, label='Regression Line')
plt.title('Linear Regression')
plt.xlabel('income')
plt.ylabel('happiness')
plt.legend()
plt.grid(True)
plt.show()
```
